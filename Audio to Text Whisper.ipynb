```python
import os
import whisper
import csv

# Path to the directory containing audio files
audio_dir = '/Users/caileyfarrell/Desktop/Deeplearning/Project/voice_extracted/KAGGLE/AUDIO/all'

# Initialize the Whisper model
model = whisper.load_model("base")

# Initialize an empty dictionary to store audio file names and their transcriptions
label_dict = {}

# Transcribe each audio file and store the transcription
for file_name in os.listdir(audio_dir):
    if file_name.endswith('.wav'):
        audio_file_path = os.path.join(audio_dir, file_name)
        
        # Transcribe the audio file
        result = model.transcribe(audio_file_path)
        
        # Store the transcription in the dictionary
        label_dict[file_name] = result['text']

# Print the dictionary to verify transcriptions
for file_name, transcription in label_dict.items():
    print(f'File: {file_name}')
    print(f'Transcription: {transcription}')
    print('-' * 80)  # Separator line for clarity

# Optionally, save the first 10 words of the transcriptions to a CSV file
with open('first_10_words_transcriptions.csv', 'w', newline='') as csvfile:
    fieldnames = ['file_name', 'first_10_words']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    for file_name, transcription in label_dict.items():
        first_10_words = ' '.join(transcription.split()[:10])
        writer.writerow({'file_name': file_name, 'first_10_words': first_10_words})

print("Transcriptions and first 10 words saved to 'first_10_words_transcriptions.csv'")
```
